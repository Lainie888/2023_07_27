```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure 建立畫布
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [5, 3, 9]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
figure = plt.figure()
axes1 = figure.add_subplot(1,3,1)
axes1.set_title("plot1")

axes2 = figure.add_subplot(1,3,2)
axes2.set_title("plot2")

axes3 = figure.add_subplot(1,3,3)
axes3.set_title("plot3")
```

```python
import matplotlib.pyplot as plt

data = [-1, -4.3, 15, 21, 31]
figure = plt.figure(figsize = (8,5))
axes = figure.add_subplot()
axes.plot(data,"rD")
```

```python
student1 = {"chinese":78, "englis":80, "math":92, "history":75, "discover":85}
student1.keys()
student1.values()

figure = plt.figure(figsize=(8,5), dpi=250, facecolor="#cccccc")
axes1 = figure.add_subplot()
axes1.plot(student1.keys(), student1.values(), "r--+", label="stu1")

student2 = {"chinese":92, "englis":60, "math":85, "history":62, "discover":73}
axes1.plot(student2.values(), "b:+", label="stu2")
axes1.legend()
plt.show()
```
