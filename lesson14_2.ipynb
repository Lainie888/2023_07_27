```python
#標準寫法
import matplotlib.pyplot as plt

# 1.製作figure 建立畫布
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [5, 3, 9]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
figure = plt.figure()
axes1 = figure.add_subplot(1,3,1)
axes1.set_title("plot1")

axes2 = figure.add_subplot(1,3,2)
axes2.set_title("plot2")

axes3 = figure.add_subplot(1,3,3)
axes3.set_title("plot3")
```

```python
import matplotlib.pyplot as plt

data = [-1, -4.3, 15, 21, 31]
figure = plt.figure(figsize = (8,5))
axes = figure.add_subplot()
axes.plot(data,"rD")
```

```python
student1 = {"chinese":78, "englis":80, "math":92, "history":75, "discover":85}
student1.keys()
student1.values()

figure = plt.figure(figsize=(8,5), dpi=250, facecolor="#cccccc")
axes1 = figure.add_subplot()
axes1.plot(student1.keys(), student1.values(), "r--+", label="stu1")

student2 = {"chinese":92, "englis":60, "math":85, "history":62, "discover":73}
axes1.plot(student2.values(), "b:+", label="stu2")
axes1.legend()
plt.show()
```

```python
#使用codespace 要安裝中文字型
#install 要加-y
#命令列:sudo apt-get update -y
#命令列:sudo apt-get install -y fonts-wqy-zenhei
#說明網址:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#字型位置:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
#mac font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#mac font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)
font1=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=12)

student1 = {"國文":78, "英文":80, "數學":92, "歷史":75, "探索":85}
figure = plt.figure(figsize=(8,5), dpi=250, facecolor="#cccccc")
axes1 = figure.add_subplot()
axes1.plot(student1.keys(), student1.values(), "r--+", label="stu1")
axes1.set_xticklabels(student1.keys(), fontproperties=font)
plt.show()
```
